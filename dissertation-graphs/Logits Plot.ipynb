```python
import torch 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np

sns.set_theme()
```

```python
probabilities = torch.tensor([0.010, 0.010, 0.15, 0.83])
logits = torch.logit(probabilities)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].set_ylim(bottom=0, top=1)
axs[0].set_xlim(left=0, right=1)
axs[0].scatter(probabilities, probabilities, marker='o', facecolor='none', edgecolors='r')

axs[1].scatter(logits, logits)
```
